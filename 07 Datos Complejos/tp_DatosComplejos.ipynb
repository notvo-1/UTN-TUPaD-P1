• Permití al usuario cargar 5 contactos con su nombre como clave y número como valor.

```python
class Agenda: #aca cree una clase porque estaba inspirado simplemente, si saco los metodos de la clase funciona y modifico el codigo un poquito anda.
    def __init__(self): #pero me salio tan bien a la primera que no quise borrarlo jaja
        self.agenda = {}
    
    def agregar_agenda(self, nombre, numero): #metodo para agregar un pana nuevo
        
        if not self.contacto_existente(nombre): 
            if self.isint(numero):           
                self.agenda[nombre] = numero
            else:
                print("Error!, ingrese un numero valido para el numero de contacto")
        else:
            print(f"El nombre ya existe en la agenda")     
    
    def contacto_existente(self, nombre): #metodo para checkear que no tengo uno con ese nombre ya
        return nombre in self.agenda
    
    def consultar(self, nombre): #devuelvo el pana y el nombre con el que se lo agendo
        if self.contacto_existente(nombre): #primero reviso si existe porque sino se rompe todo
            print(f"Contacto: {nombre} Num. Tel: {self.agenda[nombre]}")
        else:
            print(f"El contacto no existe")
    
    def isint(self, numero:str): #metodo simple para checkear si el string que ingresa el usuario tiene solo numeros, podria haber estado dentro del metodo contacto_existente solo con isdigit() pero pense que iba a usarlo de nuevo. 
        if numero.isdigit():
            return int(numero)
        return False
    
    def delete(self, nombre): #metodo para borar contacto
        print (self.agenda.pop(nombre, "El contacto no existe")) #este es simple, lo popeo usando el parametro nombre que el usuario va a pasar en el argumento de la llamada del metodo, pero si no existe no se rompe! solo devuelve "El contacto no existe", chill. 


agenda = Agenda()
while True: #bucle para hacer las veces de menu
    print("MENU")
    print("1- Agregar nuevo contacto")
    print("2- Buscar contacto")
    print("3- Eliminar contacto")
    print("0- Salir")
    opcion = input("Elija una opcion: ")
    if opcion == "1":
        print("INGRESO DE NUEVO CONTACTO")
        nombre = input("Ingrese un nombre para el contacto nuevo: ")
        numero = input("Ingrese el numero para el contacto nuevo: ")
        agenda.agregar_agenda(nombre, numero)
    elif opcion == "2":
        print("Ingrese el nombre del contacto a verificar")
        nombre = input("Ingrese un nombre para el contacto nuevo: ")
        agenda.consultar(nombre)
    elif opcion == "3":
        print("BORRAR CONTACTO")
        nombre = input("Ingrese el nombre del contacto a eliminar: ")
        agenda.delete(nombre)
        
    elif opcion == "0":
        break
    else:
        print("Opcion incorrecta, intente nuevamente")
print("Fin del programa")
        
```

5) Solicita al usuario una frase e imprime:

• Las palabras únicas (usando un set).

• Un diccionario con la cantidad de veces que aparece cada palabra.

```python
def convertir_set(frase): #convertir la frase a set para eliminar repeticiones aunque las devuelve desordenadas 
    return set(frase.replace(" ",""))

def contar_letras(frase):#contamos las letras de la frase usando un dic iterando cada letra de la frase
    cuenta = {} #init el dic
    for i in frase:#iter
        if i in cuenta:#si la letra esta en el dic le suma uno mas
            cuenta[i] += 1
        elif i == " ":#si no esta revisa si es una espacio vacio, si lo es lo ignora y pasa al sigiente elemento a iterar
            continue
        else: #si no esta en la lista y no es un espacio vacio lo agrega al dic con un valor 1
            cuenta[i] = 1
    return cuenta

frase = input("Ingrese una frase").upper()

set_palabra = convertir_set(frase)

dic_palabra = contar_letras(frase)

print(f"El set que se forma con la palabra {frase} es {sorted(set_palabra)}")#ordenamos el set
print(f"El resultado del conteo de letras que se forma con la palabra {frase} es {dic_palabra}")
```

6) Permití ingresar los nombres de 3 alumnos, y para cada uno una tupla de 3 notas. Luego, mostrá el promedio de cada alumno.

```python
n_notas = 3 #parametrice el numero de notas y alumnos para "escalar"
n_alumno = 3
notas = {} #init dic notas

for i in range(1 ,n_alumno + 1): #iteramos un rango con un for, el rango es el numero de alumnos arrancando en 1 para que no salga 0 en el print
    l_notas = [] #init list que se va a usar para almacenar las notas y lo reiniciamos por alumno
    nombre = input("Ingrese el nombre del alumno: ") #init nombre del alumno
    for j in range(1, n_notas + 1): #iteramos en un rango igual a la cantidad de notas a ingresar +1 y comenzamos en 1 por lo mismo
        nota = int(input(f"Ingrese la nota {j} del alumno numero {i}: ")) #pedimos el ingreso de la nota j para el alumno i
        l_notas.append(nota) #agregamos la nota a la lista iniciada en el nivel superiro
    notas[nombre] = tuple(l_notas) #el ejercicio pide una tupla, pero como las tuplas son inalterables primero tenemos que determinar el contenido en una lista y despues convertirla en una tupla
    
print(notas)

for nombre, calificaciones in notas.items(): #aca iteramos el dic de notas y desempaquetamos el dic con los keys en nombre y la tupla en calificaiones para procesarlas.
    suma = 0 #init suma y lo reinciamos por alumno
    for i in calificaciones: #iteramos los elementos del value que son los elementos de la tupla calificaciones desempaquetada
        suma += i #sumamos cada uno
    promedio = suma/len(calificaciones) #calculo promedio
    print(f"El promedio de notas de {nombre} es {promedio}") #print fachero

        
    
```

7) Dado dos sets de números, representando dos listas de estudiantes que aprobaron Parcial 1 y Parcial 2:

• Mostrá los que aprobaron ambos parciales.

• Mostrá los que aprobaron solo uno de los dos.

• Mostrá la lista total de estudiantes que aprobaron al menos un parcial (sin repetir).

```python
from random import randint #import randint para usar en los sets

parcial_1 = set([randint(1,100) for _ in range(10)]) #determinamos 2 sets con numeros randoms en un rango para obtener distintos resultados en cada prueba. 
parcial_2 = set([randint(1,100) for _ in range(10)]) #uso list comprehension porque se ve copado y creo que usa menos memoria, no recuerdo. 

print("Lista aprobador Parcial 1:",parcial_1)
print("Lista aprobador Parcial 1:",parcial_2)

alumno_aprobaron_todo = parcial_1.intersection(parcial_2) #metodos de los sets para resolver los problemas. 
print(f"{len(alumno_aprobaron_todo)} son los alumnos que aprobaron los dos examenes {alumno_aprobaron_todo}")

alumno_aprobaron_solo_uno = parcial_1.symmetric_difference(parcial_2)
print(f"{len(alumno_aprobaron_solo_uno)} son los alumnos que solo aprobaron 1 parcial {alumno_aprobaron_solo_uno}")

alumno_aprobaron_al_menos_uno = parcial_1.union(parcial_2)
print(f"{len(alumno_aprobaron_al_menos_uno)} son los alumnos que al menos aprobaron un parcial {alumno_aprobaron_al_menos_uno}")

"""Esta bueno porque da resultados distintos cada vez que lo ejecuto"""


```

9) Creá una agenda donde las claves sean tuplas de (día, hora) y los valores sean eventos.


```python
def agregar_evento(agenda:dict, fecha:tuple, evento:str):
    if es_fecha(fecha):
        if fecha not in agenda:
            agenda[fecha] = evento
            print(f"Evento '{evento}' agregado para el {fecha[0]} a las {fecha[1]}.")
            return True
        else:
            print(f"Ya hay un evento agendado para el dia {fecha[0]} a las {fecha[1]}")
    else:
        print("Error! fecha duplicada o invalida")
    return False

def es_fecha(tupla:tuple):
    if not isinstance(tupla, tuple):
        return False
    dias = ["domingo", "lunes", "martes", "miercoles", "jueves", "viernes", "sabado"]
    dia, hora = tupla
    if dia in dias and es_hora(hora):
        return True
    else: 
        return False
    
def es_hora(hora:str):
    if ":" not in hora:
        return False
    if hora.count(":") != 1:
        return False


    h, m = hora.split(":")
    if not h.isdigit() or not m.isdigit():
        return False
    h = int(h)
    m = int(m)

    if h >=0 and h <24 and m >= 0 and m < 60:
        return True

    return False       


def consultar_evento(agenda:dict, fecha:tuple):
    if es_fecha(fecha) and fecha in agenda:
        print(f"Para el dia {fecha[0]} a las {fecha[1]} hay agendado {agenda[fecha]}.")
    else:
        print("No hay nada registrado para esa fecha")

def eliminar_evento(agenda:dict, fecha:tuple):
    if es_fecha(fecha):
        eliminado = agenda.pop(fecha, None)
        if eliminado is None:
            print("La fecha no existe")
        else:
            print(f"Evento elminado: {eliminado}")
    else:
        print("Fecha invalida")

agenda = {}

while True:
    print(f"La agenda va asi {agenda}")
    print("AGENDA DE EVENTOS")
    print("1 - Agregar evento")
    print("2 - Consultar evento")
    print("3 - Eliminar evento")
    print("0 - Salir")

    opcion = input("Ingrese una opción: ")

    if opcion not in ["1", "2", "3", "0"]:
        print("Opción incorrecta")
    else:
        if opcion == "1":
            dia = input("Ingrese el día (ej: lunes): ").lower()
            hora = input("Ingrese la hora (formato HH:MM): ")
            evento = input("Ingrese la descripción del evento: ")
            fecha = (dia, hora)
            agregar_evento(agenda, fecha, evento)

        elif opcion == "2":
            dia = input("Ingrese el día: ").lower()
            hora = input("Ingrese la hora: ")
            fecha = (dia, hora)
            consultar_evento(agenda, fecha)

        elif opcion == "3":
            dia = input("Ingrese el día: ").lower()
            hora = input("Ingrese la hora: ")
            fecha = (dia, hora)
            eliminar_evento(agenda, fecha)

        elif opcion == "0":
            print("Saliendo del programa...")
            break
```
